```python
df = df.drop_duplicates()
```

```python
numeric_cols = [
    "temperature_celsius", "feels_like_celsius", "humidity", "wind_kph",
    "gust_kph", "pressure_mb", "precip_mm", "cloud", "visibility_km", "uv_index",
    "air_quality_Carbon_Monoxide", "air_quality_Ozone", "air_quality_Nitrogen_dioxide",
    "air_quality_Sulphur_dioxide", "air_quality_PM2.5", "air_quality_PM10",
    "air_quality_us-epa-index", "air_quality_gb-defra-index"
]
```

```python
print("Basic Statistics:")
display(df[numeric_cols].describe())
```

```python
from scipy.stats import skew, kurtosis

print("Skewness & Kurtosis:\n")
for col in numeric_cols:
    col_skew = skew(df[col].dropna())
    col_kurt = kurtosis(df[col].dropna())
    print(f"{col:<35} Skew = {col_skew:>7.2f}    Kurtosis = {col_kurt:>7.2f}")
```

```python
stats = []
for col in numeric_cols:
    stats.append([col, skew(df[col].dropna()), kurtosis(df[col].dropna())])

df_stats = pd.DataFrame(stats, columns=["Variable", "Skewness", "Kurtosis"])


plt.figure(figsize=(10,6))
sns.heatmap(
    df_stats.set_index("Variable"),
    annot=True,
    cmap="RdYlGn",      #
    center=0,
    fmt=".1f"
)
plt.title("Skewness & Kurtosis of Numeric Features", fontsize=14, weight="bold")
plt.show()
```

```python
from scipy.stats import zscore
import numpy as np

# Make a copy of the dataset
df_cleaned = df.copy()

# Select columns to clean
cols_to_clean = ["pressure_mb", "air_quality_Sulphur_dioxide"]

for col in cols_to_clean:
    col_data = df_cleaned[col].dropna()
    z_scores = zscore(col_data)

    # Mask: True for values within 3 std deviations
    mask = abs(z_scores) < 3

    # Replace outliers with NaN (only within valid indices)
    df_cleaned.loc[col_data.index, col] = col_data.where(mask, np.nan)

# Compare before vs after
print("Before cleaning:")
print(df[cols_to_clean].describe())

print("\nAfter cleaning:")
print(df_cleaned[cols_to_clean].describe())
```

```python
from scipy.stats import skew, kurtosis

print("\nSkewness & Kurtosis after cleaning:\n")
for col in numeric_cols:

    if col in df_cleaned.columns:
        col_data = df_cleaned[col].dropna()
        if len(col_data) > 0:
            col_skew = skew(col_data)
            col_kurt = kurtosis(col_data)
            print(f"{col:<35} Skew = {col_skew:>7.2f}   Kurtosis = {col_kurt:>7.2f}")
        else:
            print(f"{col:<35} Skew =  N/A     Kurtosis =  N/A (no data)")
    else:
        print(f"{col:<35} Not found in df_cleaned")
```

```python
stats = []
for col in numeric_cols:
    stats.append([col, skew(df_cleaned [col].dropna()), kurtosis(df_cleaned [col].dropna())])

df_stats = pd.DataFrame(stats, columns=["Variable", "Skewness", "Kurtosis"])

plt.figure(figsize=(10,6))
sns.heatmap(
    df_stats.set_index("Variable"),
    annot=True,
    cmap="RdYlGn",
    center=0,
    fmt=".1f"
)
plt.title("Skewness & Kurtosis of Numeric Features", fontsize=14, weight="bold")
plt.show()
```

```python

plt.figure(figsize=(16, 12))
for i, col in enumerate(numeric_cols[:9], 1):
    plt.subplot(3, 3, i)
    sns.histplot(df_cleaned[col].dropna(), kde=True, bins=30, color="seagreen")
    plt.title(f"Distribution of {col}")
plt.tight_layout()
plt.show()
```

```python
df_cleaned["last_updated"] = pd.to_datetime(df_cleaned["last_updated"])
df_cleaned["month"] = df_cleaned["last_updated"].dt.month
cols_to_analyze = [
    "temperature_celsius",
    "precip_mm",
    "humidity",
    "wind_kph",
    "air_quality_PM2.5",
    "air_quality_PM10",
    "uv_index"
]
monthly_avg = df_cleaned.groupby("month")[cols_to_analyze].mean()
monthly_avg.plot(kind="line", marker="o", figsize=(12, 7))
plt.title("Seasonal Patterns: Weather & Air Quality Indicators")
plt.xlabel("Month")
plt.ylabel("Average Value")
plt.legend(title="Indicators")
plt.grid(True)
plt.show()
```

```python
import plotly.express as px
from scipy.stats import zscore
import numpy as np


df_cleaned = df.copy()


cols_to_clean = ["pressure_mb", "air_quality_Sulphur_dioxide"]

for col in cols_to_clean:
    col_data = df_cleaned[col].dropna()
    z_scores = zscore(col_data)
    mask = abs(z_scores) < 3
    df_cleaned.loc[col_data.index, col] = col_data.where(mask, np.nan)
country_avg_temp = df_cleaned.groupby('country')['temperature_celsius'].mean().reset_index()
fig = px.choropleth(
    country_avg_temp,
    locations="country",
    locationmode="country names",
    color="temperature_celsius",
    hover_name="country",
    color_continuous_scale="Viridis",
    title="Average Temperature by Country"
)

fig.show()
```

```python
avg_humidity_by_wind = df_cleaned.groupby('wind_direction')['humidity'].mean().reset_index()
plt.figure(figsize=(12, 6))
sns.barplot(x='wind_direction', y='humidity', data=avg_humidity_by_wind, palette='Blues', hue='wind_direction', legend=False)
plt.title('Average Humidity by Wind Direction')
plt.xlabel('Wind Direction')
plt.ylabel('Average Humidity (%)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='temperature_celsius', y='humidity', data=df_cleaned)
plt.title('Temperature vs. Humidity')
plt.xlabel('Temperature (C)')
plt.ylabel('Humidity (%)')
plt.grid(True)
plt.show()
```

```python
first_10_countries = df_cleaned['country'].unique()[:10]

# Filter the DataFrame to include only data from the first 10 countries
df_first_10_countries = df_cleaned[df_cleaned['country'].isin(first_10_countries)]

# Create the box plot
plt.figure(figsize=(14, 7))
sns.boxplot(x='country', y='temperature_celsius', data=df_first_10_countries, palette='viridis')
plt.title('Temperature Distribution for the First 10 Countries')
plt.xlabel('Country')
plt.ylabel('Temperature (C)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

## Interactive Dashboard Wireframe Layout

Here is a possible wireframe layout for the interactive weather and air quality dashboard:

**Overall Layout:**

*   **Header:** Title of the dashboard (e.g., "Global Weather and Air Quality Dashboard")
*   **Sidebar (Left):**
    *   Filters/Controls:
        *   Country selection (Dropdown or multi-select)
        *   Date/Time range slider
        *   Metric selection (e.g., Temperature, Humidity, PM2.5)
        *   Visualization type selection (e.g., Scatter plot, Bar chart, Histogram)
*   **Main Content Area (Right):** Divided into sections for different visualizations.

**Main Content Sections:**

*   **Section 1: Geographical Overview (Top)**
    *   Choropleth map showing the selected metric (e.g., Average Temperature) by country.
*   **Section 2: Key Metrics Summary (Middle Left)**
    *   Cards or small plots showing key summary statistics for the selected metric (e.g., Current Temperature, Average Humidity, PM2.5 Level).
*   **Section 3: Trend over Time (Middle Right)**
    *   Line plot showing the trend of the selected metric over the selected date/time range.
*   **Section 4: Distribution Analysis (Bottom Left)**
    *   Histogram or Box plot showing the distribution of the selected metric for the selected country/region.
*   **Section 5: Relationship between Metrics (Bottom Right)**
    *   Scatter plot or Heatmap showing the relationship between two selected metrics.

**Interactive Elements:**

*   Selecting a country from the sidebar filter updates all visualizations to show data for that country.
*   Adjusting the date/time range slider updates the trend over time plot and summary statistics.
*   Selecting a different metric updates the choropleth map, trend plot, summary statistics, and distribution plot.
*   Hovering over data points on plots displays detailed information (tooltips).

This is a basic wireframe and can be adjusted based on the specific metrics and visualizations you want to prioritize and the level of interactivity desired.
