```python
import pandas as pd

df = pd.read_csv("GlobalWeatherRepository.csv")

print("Dataset shape:", df.shape)
print("\nColumn names:", df.columns.tolist())
print("\nData types:")
print(df.dtypes)
print("\nQuick summary:")
print(df.info())
print("\nSample rows:")
print(df.head())
```

```python
missing_summary = df.isnull().sum().sort_values(ascending=False)
print("\nMissing values per column:\n", missing_summary)
```

```python
print("\nNumber of duplicate rows:", df.duplicated().sum())
```

```python
print("\nSummary stats:")
print(df.describe(include="all"))
```

```python
import numpy as np

for col in df.columns:
    if df[col].dtype in [np.float64, np.int64]:
        df[col].fillna(df[col].mean(), inplace=True)
    else:
        df[col].fillna(df[col].mode()[0], inplace=True)


df.drop_duplicates(inplace=True)
```

```python
if 'temperature_F' in df.columns:
    df['temperature_C'] = (df['temperature_F'] - 32) * 5/9
```

```python
if 'precipitation_mm' in df.columns:
    df['precipitation_cm'] = df['precipitation_mm'] / 10
```

```python
numeric_cols = df.select_dtypes(include=[np.number]).columns
df[numeric_cols] = (df[numeric_cols] - df[numeric_cols].min()) / (df[numeric_cols].max() - df[numeric_cols].min())
```

```python
if 'date' in df.columns:
    df['date'] = pd.to_datetime(df['date'], errors='coerce')

    # Group by year-month
    monthly_avg = df.groupby(df['date'].dt.to_period("M")).mean(numeric_only=True)
    monthly_avg = monthly_avg.reset_index()

    print("\nMonthly average sample:\n", monthly_avg.head())
```

```python
df = pd.read_csv("GlobalWeatherRepository.csv")

df['last_updated'] = pd.to_datetime(df['last_updated'], errors='coerce')


df = df.dropna(subset=['last_updated'])


df.set_index('last_updated', inplace=True)


monthly_avg = df.resample('M').mean(numeric_only=True)


monthly_avg = monthly_avg.reset_index()


print(monthly_avg.head())
```

```python
import pandas as pd
import numpy as np


df = pd.read_csv("GlobalWeatherRepository.csv")


print("Columns in dataset:\n", df.columns.tolist())


if 'temperature_F' in df.columns:
    df['temperature_C'] = (df['temperature_F'] - 32) * 5/9
    print("\n✅ Converted 'temperature_F' to 'temperature_C'")


if 'precipitation_mm' in df.columns:
    df['precipitation_cm'] = df['precipitation_mm'] / 10
    print("✅ Converted 'precipitation_mm' to 'precipitation_cm'")


if 'wind_speed_kmh' in df.columns:
    df['wind_speed_ms'] = df['wind_speed_kmh'] / 3.6
    print("✅ Converted 'wind_speed_kmh' to 'wind_speed_ms'")


numeric_cols = df.select_dtypes(include=[np.number]).columns


df_normalized = df.copy()
df_normalized[numeric_cols] = (
    df[numeric_cols] - df[numeric_cols].min()
) / (df[numeric_cols].max() - df[numeric_cols].min())

print("\n✅ Numeric values normalized (range 0–1)")


print("\nConverted and normalized data preview:")
print(df_normalized.head())
```
