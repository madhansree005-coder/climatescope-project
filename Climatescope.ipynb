```python
import os
import pandas as pd
```

```python
print("Working dir:", os.getcwd())
print("Files here:", os.listdir('.')[:20])
```

```python
paths = [
    "GlobalWeatherRepository.csv"
]

csv_path = None
for p in paths:
    if os.path.exists(p):
        csv_path = p
        break

if csv_path is None:
    raise FileNotFoundError("CSV not found. Upload the file to Colab or set the correct path.")
```

```python
df = pd.read_csv(csv_path)
```

```python
df.columns = df.columns.str.strip()
print("Column names (first 20):", df.columns.tolist()[:20])
```

```python
first_col = df.columns[0]
print("\nFirst column name:", first_col)
print(df[first_col].head())

if 'country' in df.columns:
    print("\n'country' column head:")
    print(df['country'].head())
else:
    print("\nNo column named 'country' exactly — check the printed column list above.")
```
